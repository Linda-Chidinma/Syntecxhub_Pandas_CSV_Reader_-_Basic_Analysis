### Student performance data

A sample dataset was created to demonstrate Pandas CSV reading and basic analysis techniques.

```python
# Import Libraries
import pandas as pd
```

```python
# Create a Sample Dataset 
data = {
    "Student_ID": [101, 102, 103, 104, 105],
    "Name": ["Ayo", "Bola", "Chidi", "Dami", "Efe"],
    "Score": [78, 85, 62, 90, 70],
    "Age": [15, 16, 15, 17, 16]
}

df = pd.DataFrame(data)
df
```

```python
# Save Dataset to CSV 
df.to_csv("students.csv", index=False)
```

```python
# Read CSV into DataFrame 
students_df = pd.read_csv("students.csv")
students_df
```

```python
# Inspect Data
students_df.head()
```

```python
students_df.tail()
```

```python
students_df.dtypes
```

```python
# Summary Statistics 
students_df.describe()
```

```python
print(students_df["Score"].mean())
print(students_df["Score"].median())
print(students_df["Score"].min())
print(students_df["Score"].max())
print(students_df["Score"].count()
```

```python
# Filter Rows
high_scores = students_df[students_df["Score"] > 75]
high_scores
```

```python
# Select Columns
name_and_score = students_df[["Name", "Score"]]
name_and_score
```

```python
# Slice Subsets
students_df.iloc[0:3]
```

```python
# Save Filtered Results 
high_scores.to_csv("high_scoring_students.csv", index=False)
```
